# Create an If-Statement Function

```python
# Create a function that groups rows based on their price
def price_label(row):

    if row['prices'] <= 5:
        return 'Low-range prodcut'
    elif (row['prices'] > 5) and (row['prices'] <= 15):
        return 'Mid-range product'
    elif row['prices'] > 15:
        return 'High range'
    else: return 'Not enough data'
```

# If-Statements with For-Loops

```python
# Find out which day most orders take place
df_ords_prods_merge['orders_day_of_week'].value_counts(dropna=False)
```

```python
# Create a for-loop that classifies the orders_day_of_week column into different categories
result = []

for value in df_ords_prods_merge["orders_day_of_week"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

### 2. Suppose your clients have changed their minds about the labels you created in your â€œbusiest_dayâ€ column. Now, they want â€œBusiest dayâ€ to become â€œBusiest daysâ€ (plural). This label should correspond with the two busiest days of the week as opposed to the single busiest day. At the same time, theyâ€™d also like to know the two slowest days. Create a new column for this using a suitable method.

```python
# Find out which days are busiests and least busy
df_ords_prods_merge['orders_day_of_week'].value_counts(dropna=False)
```

```python
# Create a for-loop that classifies the orders_day_of_week column into the necessary categories
result = []

for value in df_ords_prods_merge["orders_day_of_week"]:
  if value == 0 or value == 1:
    result.append("Two Busiest days")
  elif value == 4 or value == 3:
    result.append("Two slowest days")
  else:
    result.append("Regularly busy")
```

```python
# Create a new column in df_ords_prods_merge where the for-loop results will be placed
df_ords_prods_merge['busiest_days'] = result
```

```python
# Check the value counts of the busiest_days column
df_ords_prods_merge['busiest_days'].value_counts(dropna=False)
```

#### The value for the two busiest days match the sum of the count of 0 and 1 in the orders_day_of_week column. The same is true for the value of the two slowest days. So, the for-loop appears to have accurately classified the two busiest days, the two slowest days, and all other days into different categories in the busiest_days column.

## 4. When too many users make Instacart orders at the same time, the app freezes. The senior technical officer at Instacart wants you to identify the busiest hours of the day. Rather than by hour, they want periods of time labeled â€œMost orders,â€ â€œAverage orders,â€ and â€œFewest orders.â€ Create a new column containing these labels called â€œbusiest_period_of_day.â€

```python
# First, I need a value count of each hour of the day
df_ords_prods_merge['order_hour_of_day'].value_counts(dropna=False)
```

```python
# Create a for-loop that classifies the order hours into 3 different categories based on their number of orders
result_hours = []

for value in df_ords_prods_merge["order_hour_of_day"]:
  if value in [10,11,14,15,13,12,16,9,17]:
    result_hours.append("Most orders")
  elif value in [1,5,2,4,3]:
    result_hours.append("Fewest orders")
  else:
    result_hours.append("Average orders")
```

##### # I'm grouping the hours based on obvious gaps in the number of orders, such as the gap between 115700 orders at 1 and 218769 at 0.

```python
# Create the new column for this loop
df_ords_prods_merge['busiest_period_of_day'] = result_hours
```
