# Sampling Data

```python
# Creating a list of random numbers
# Comparing the list to 0.7 to see if each value is higher or lower than 0.7
# Storing the results of the comparison in a new list, 'dev'
np.random.seed(4)
dev = np.random.rand(len(ords_prods_merge)) <= 0.7
```

```python
# Store 70% of the sample in the df 'big'
# Store 30% of the sample in the df 'small'
big = ords_prods_merge[dev]
small = ords_prods_merge[~dev]
```

```python
# Checking the length of the enitre ords_prods_merge df
len(ords_prods_merge)
```

```python
# Checking the length of the big and small dfs
len(big) + len(small)
```

```python
# Create a smaller df that only contains two columns from the small df
df_2=small[['orders_day_of_week', 'prices']]
```

# Creating Line Charts

```python
# Create a line chart of the orders_day_oif_week and prices columns using the small sample
line = sns.lineplot(data = df_2, x = 'orders_day_of_week', y = 'prices')
```
