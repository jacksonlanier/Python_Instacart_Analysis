# You need to provide the Instacart senior stakeholders with descriptive findings about sales. Create a histogram of the “order_hour_of_day” column.

```python
# Visualize the department names with a bar chart using abbreviated data points
def abbreviate_count(count):
    if count >= 1000000:
        return f"{float(count) / 1000000:.1f}m"
    elif count >= 1000:
        return f"{float(count) / 1000:.1f}k"
    elif count >= 100:
        return f"{count // 100}h"
    else:
        return str(count)


# Create a histogram of the order_hour_of_day column
hist_ohod = ords_prods_cust['order_hour_of_day'].plot.hist(bins=24, edgecolor='black', figsize=(12,8))

# Access the current Axes object
ax = plt.gca()

# Add data labels to the bars
ax.bar_label(ax.containers[0], fmt=abbreviate_count, fontsize='7')

# Other customizations
hist_ohod.ticklabel_format(axis='y', style='plain')
hist_ohod.set_xlabel('Hour of Day')
hist_ohod.set_title('Order Frequency per Hour of Day')
```

```python
# Create a histogram of the order_hour_of_day column
hist_ohod = ords_prods_cust['order_hour_of_day'].plot.hist(bins = 24, edgecolor='black')
hist_ohod.ticklabel_format(axis='y', style='plain')
hist_ohod.set_xlabel('Hour of Day')
hist_ohod.set_title('Order Frequency per Hour of Day')
```

### There aren't many orders from 1:00 to 5:00. # The marketing team is curious about the distribution of orders among customers in terms of loyalty. Create a bar chart from the “loyalty_flag” column.

```python
# Visualize with a bar chart using abbreviated data points
def abbreviate_count(count):
    if count >= 1000000:
        return f"{float(count) / 1000000:.1f}m"
    elif count >= 1000:
        return f"{count // 1000}k"
    elif count >= 100:
        return f"{count // 100}h"
    else:
        return str(count)



# Count the values
loyal = ords_prods_cust['loyalty_flag'].value_counts(dropna=False)

#Plot
bar_loyal = loyal.plot.bar(color =['royalblue', 'blueviolet', 'orange'], rot=0)

# Access the current Axes object
ax = plt.gca()

for i, container in enumerate(ax.containers):
    if i % 1 == 0: 
        ax.bar_label(container, fmt=abbreviate_count, fontsize='8')



# Customize
bar_loyal.ticklabel_format(axis='y', style='plain')
bar_loyal.set_title('Order Count Based on User Loyalty')
bar_loyal.set_ylabel('Orders')
bar_loyal.set_xlabel('User Loyaltly', weight='bold')
```

```python
# Exporting the bar chart to my visualization folder
bar_loyal.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'bar_loyal.png'), bbox_inches='tight')
```

# Check whether there’s a difference in expenditure (the “prices” column) depending on the hour of the day. (Hint: To check this, you need to use an accurate sample for your line chart!)
